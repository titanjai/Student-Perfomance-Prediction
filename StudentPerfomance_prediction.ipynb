```python
import pandas as pd #dataset reading
df=pd.read_csv("student_data.csv")
print(df.head())
```

```python
print(df.columns) #all columns
```

```python
print(df.shape) #dataset info
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in df.columns:   #label encoding
  if df[col].dtype=="object":
    df[col]=le.fit_transform(df[col])

print(df.head())
```

```python
x=df.drop("G3",axis=1) #input variables
y=df["G3"]  #target variable
```

```python
from sklearn.model_selection import train_test_split   #train test data split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression  #assigning model
model=LinearRegression()
model.fit(X_train,Y_train)   #training
print("Training completed")
print("Training Score:",model.score(X_train,Y_train))
print("Testing Score:",model.score(X_test,Y_test))
```

```python
y_pred=model.predict(X_test) #prediction
print(y_pred[:5])
```

```python
from sklearn.metrics import mean_absolute_error #metrics score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np

mae=mean_absolute_error(Y_test,y_pred)
mse=mean_squared_error(Y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(Y_test,y_pred)

print("Mean Absolute Error: ",mae)
print("Mean Squared Error: ",mse)
print("Root Mean Squared Error: ",rmse)
print("Mean Squared Error: ",rmse)

print("R2 Score: ",r2)
```
